```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

Example of a figure, save as png, svg and pdf

```python
# generate a random data set of 100 values.  The 100 values are actually governed by a standard normal
# distribution, with a mean of 0 and a standard deviation of 1.
x = np.random.normal(size=100)
```

```python
plt.rcParams['figure.figsize'] = (13.0,6.0) # set default figure size a bit bigger for all figures
sns.histplot(x, kde=True);
plt.savefig('../figures/figure-example-01.png', dpi=600)
plt.savefig('../figures/figure-example-01.svg', dpi=600)
plt.savefig('../figures/figure-example-01.pdf', dpi=600)
```

Example of generating a LaTeX table from numpy array.

```python
df = pd.DataFrame(np.random.random((5, 5)))
df.columns = ['Attribute 1', 'Another Attribute', 'Data', 'Input', 'Output']
df
```

```python
print(df.style.to_latex())
```
